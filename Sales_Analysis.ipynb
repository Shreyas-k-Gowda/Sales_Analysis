```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Sales Data.csv')
```

```python
#Processing the Data
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df['Day'] = df['Order Date'].dt.day
df['Weekday'] = df['Order Date'].dt.day_name()

sns.set(style="whitegrid")
```

```python
# This plot shows Sales Over Time
plt.figure(figsize=(14, 5))
df.groupby('Order Date')['Sales'].sum().plot()
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot shows Sales by Region
plt.figure(figsize=(8,5))
df.groupby('Region')['Sales'].sum().plot(kind='bar', color='skyblue')
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Plot shows Sales by Category
plt.figure(figsize=(6,4))
df.groupby('Category')['Sales'].sum().plot(kind='bar', color='orange')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Sales by Sub-Category
plt.figure(figsize=(12,5))
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='bar', color='green')
plt.title("Sales by Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#5. Profit by Segment
plt.figure(figsize=(6,4))
df.groupby('Segment')['Profit'].sum().plot(kind='bar', color='purple')
plt.title("Profit by Segment")
plt.ylabel("Profit")
plt.show()
```

```python
# 6.Avg Discount by Sub-Category
plt.figure(figsize=(10,4))
df.groupby('Sub-Category')['Discount'].mean().sort_values().plot(kind='bar', color='salmon')
plt.title("Average Discount by Sub-Category")
plt.ylabel("Average Discount")
plt.xticks(rotation=45)
plt.show()
```

```python
#  7. Sales vs Quantity
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Quantity', y='Sales')
plt.title("Sales vs Quantity")
plt.show()
```

```python

# 8. Sales vs Profit
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Profit', y='Sales')
plt.title("Sales vs Profit")
plt.show()
```

```python

#  9. Monthly Sales Trend
plt.figure(figsize=(8,5))
df.groupby('Month')['Sales'].sum().plot(kind='line', marker='o', color='teal')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
#10. Weekday Order Volume
plt.figure(figsize=(8,5))
df['Weekday'].value_counts().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
).plot(kind='bar', color='coral')
plt.title("Orders by Day of Week")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Heatmap of Correlation
plt.figure(figsize=(8,6))
sns.heatmap(df[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#  12. Sub-Category vs Profit Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(data=df, x='Sub-Category', y='Profit', palette='Set2')
plt.title("Profit Distribution by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 13. Ship Mode vs Sales
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Ship Mode', y='Sales')
plt.title("Sales Distribution by Ship Mode")
plt.xticks(rotation=45)
plt.show()
```

```python
#14. Count of Orders by Segment
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Segment', palette='Set1')
plt.title("Order Count by Segment")
plt.show()
```

```python
# 15. Pie Chart of Segment Distribution
plt.figure(figsize=(6,6))
df['Segment'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.title("Segment Distribution")
plt.ylabel("")
plt.show()
```

```python
# 16. Sales by Year
plt.figure(figsize=(6,4))
df.groupby('Year')['Sales'].sum().plot(kind='bar', color='olive')
plt.title("Sales by Year")
plt.ylabel("Sales")
plt.show()
```

```python
# 17. Profit Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Profit'], bins=50, kde=True, color='tomato')
plt.title("Distribution of Profit")
plt.show()
```

```python
# 18. Sales vs Discount
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Sales', hue='Category')
plt.title("Sales vs Discount by Category")
plt.show()
```

```python
# 19. Top 10 Cities by Sales
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_cities.plot(kind='bar', color='darkcyan')
plt.title("Top 10 Cities by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 20. Top 10 Customers by Profit
top_customers = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_customers.plot(kind='bar', color='seagreen')
plt.title("Top 10 Customers by Profit")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# Load and parse dates
df = pd.read_csv('Sales Data.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month-Year'] = df['Order Date'].dt.to_period('M').astype(str)

# Group by Month-Year, Segment, Category
monthly_profit = df.groupby(['Month-Year', 'Segment', 'Category'])['Profit'].sum().reset_index()

# Pivot the data for heatmap
profit_pivot = monthly_profit.pivot_table(index='Month-Year', columns=['Segment', 'Category'], values='Profit', fill_value=0)

# Smooth values using rolling mean (optional)
profit_pivot_smooth = profit_pivot.rolling(window=2, min_periods=1).mean()
```

```python

# Plot the heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(profit_pivot_smooth.T, cmap='RdYlGn', center=0, annot=False, fmt='.0f', linewidths=.5, cbar_kws={'label': 'Profit'})
plt.title(' Monthly Profit Trends by Segment & Category (Smoothed)')
plt.xlabel('Month')
plt.ylabel('Segment / Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
# Group by Year, Month, and Category
monthly_sales_category = df.groupby([df['Order Date'].dt.to_period('M').astype(str), 'Category'])['Sales'].sum().reset_index()

# Pivot the data for the heatmap
sales_pivot = monthly_sales_category.pivot_table(index='Order Date', columns='Category', values='Sales', fill_value=0)

```

```python
# Plot the heatmap
plt.figure(figsize=(16, 7))
sns.heatmap(sales_pivot.T, cmap='YlGnBu', annot=False, linewidths=.5, cbar_kws={'label': 'Total Sales'})
plt.title('Monthly Sales by Category Heatmap')
plt.xlabel('Month')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
